```python
# prompt: load one large data set from keras (not fashion_inst) and get and print its description from the source

import tensorflow as tf
import tensorflow_datasets as tfds

# Load one large dataset from Keras
# We'll use the 'cifar100' dataset which is larger than 'mnist' or 'fashion_mnist'
# You can explore other datasets using tfds.list_builders()
dataset_name = 'cifar100'
(train_ds, test_ds), ds_info = tfds.load(
    dataset_name,
    split=['train', 'test'],
    with_info=True,
    as_supervised=True,
)

# Get and print the description from the source
print(ds_info.description)
print("\nDataset Features:")
print(ds_info.features)
print("\nDataset Splits:")
print(ds_info.splits)
print("\nDataset Number of Examples:")
print(ds_info.splits['train'].num_examples)
print(ds_info.splits['test'].num_examples)
```

```python
# prompt: show a sample data from the imported dataset

# Take one sample from the training dataset
sample_example = next(iter(train_ds))

# The sample is a tuple of (image, label)
image, label = sample_example

# Print the shape and data type of the image
print("\nSample Image Shape:", image.shape)
print("Sample Image Data Type:", image.dtype)

# Print the label (it's an integer representing the class)
print("Sample Label:", label.numpy())

# You can also visualize the image if matplotlib is installed
# !pip install matplotlib
import matplotlib.pyplot as plt
plt.imshow(image.numpy())
plt.axis('off')
plt.show()
```

```python
# prompt: make a neural network to predict the images.

# Normalize images to the range [0, 1]
def normalize_img(image, label):
    """Normalizes images: `uint8` -> `float32`."""
    return tf.cast(image, tf.float32) / 255., label

# Apply normalization and shuffle and batch the datasets
batch_size = 32

train_ds = train_ds.map(normalize_img, num_parallel_calls=tf.data.AUTOTUNE)
train_ds = train_ds.cache()
train_ds = train_ds.shuffle(ds_info.splits['train'].num_examples)
train_ds = train_ds.batch(batch_size)
train_ds = train_ds.prefetch(tf.data.AUTOTUNE)

test_ds = test_ds.map(normalize_img, num_parallel_calls=tf.data.AUTOTUNE)
test_ds = test_ds.cache()
test_ds = test_ds.batch(batch_size)
test_ds = test_ds.prefetch(tf.data.AUTOTUNE)

# Define the model architecture
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(ds_info.features['label'].num_classes) # Output layer with number of classes
])

# Compile the model
model.compile(
    optimizer=tf.keras.optimizers.Adam(0.001),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)

# Print the model summary
model.summary()

# Train the model
epochs = 10 # You can adjust the number of epochs
history = model.fit(
    train_ds,
    epochs=epochs,
    validation_data=test_ds
)

# Evaluate the model on the test dataset
loss, accuracy = model.evaluate(test_ds)
print(f"\nTest Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")

# Plot training history
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Training and Validation Loss')

plt.show()
```

```python
# prompt: predict 10 random pictures and print the report and show the metrics.

import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Get 10 random images from the test dataset
test_ds_for_prediction = test_ds.unbatch().take(10)
images_to_predict = []
true_labels = []
for image, label in test_ds_for_prediction.as_numpy_iterator():
    images_to_predict.append(image)
    true_labels.append(label)

images_to_predict = np.array(images_to_predict)
true_labels = np.array(true_labels)

# Make predictions
predictions = model.predict(images_to_predict)
predicted_labels = np.argmax(predictions, axis=1)

# Print prediction report
print("\n--- Prediction Report ---")
for i in range(len(images_to_predict)):
    true_class_name = ds_info.features['label'].int2str(true_labels[i])
    predicted_class_name = ds_info.features['label'].int2str(predicted_labels[i])
    print(f"Image {i+1}: True Label: {true_class_name}, Predicted Label: {predicted_class_name}")
    # Optionally, display the image
    # plt.imshow(images_to_predict[i])
    # plt.title(f"True: {true_class_name}, Predicted: {predicted_class_name}")
    # plt.axis('off')
    # plt.show()

# Print classification report for the 10 images
print("\n--- Classification Report (for 10 random samples) ---")
# Since we are only predicting on 10 samples, the report will be very limited.
# For a meaningful report, you'd usually predict on a larger subset or the entire test set.
# We'll create a dummy target_names list covering all possible classes for completeness
target_names = [ds_info.features['label'].int2str(i) for i in range(ds_info.features['label'].num_classes)]

# Filter target_names to only include labels present in the true_labels and predicted_labels
unique_labels = np.unique(np.concatenate((true_labels, predicted_labels)))
filtered_target_names = [target_names[i] for i in unique_labels]

print(classification_report(true_labels, predicted_labels, labels=unique_labels, target_names=filtered_target_names, zero_division=0))

# Show confusion matrix (for 10 images, this will be very sparse)
print("\n--- Confusion Matrix (for 10 random samples) ---")
conf_matrix = confusion_matrix(true_labels, predicted_labels, labels=unique_labels)

plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=filtered_target_names, yticklabels=filtered_target_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix (10 Random Samples)')
plt.show()
```
